## Exercice 1 - Manipuler des données bioinformatiques (7 points)

A la suite d'une analyse protéomique nous disposons d'une liste de protéines. Chaque protéine est représentée par une liste contenant son nom, sa séquence d'acides aminés (simplifiée ici par une chaîne de caractères) et sa fonction biologique principale.

Objectif : Transformer cette liste de protéines en un DataFrame Pandas pour faciliter les analyses fonctionnelles.

````python
proteines = [
    ["Insuline", "MALWMRLLPLLALLALWGPDPAAFVNQHLCGSHLVEALYLVCGERGFFYTPKA", "Hormone"],
    ["Hémoglobine", "VLSPADKTNVKAAWGKVGAHAGKKFETLEKFDNKDKIAALGLESELEAEMK", "Transport"],
    ["Enzyme X", "MKQWEDLKKAGDKEYKLVDFGHPQETLVRYSRVK", "Enzyme"],
    ["Récepteur Y", "AVFPQPELTLKWMLGIVFTWICFLVFLLVYGVILVI", "Récepteur"],
    ["Protéine structurale Z", "GVGVPGEVEMKLGYIGAHFGGEAG", "Structure"],
    ["Enzyme Y", "DVQLQESGPGLVKPSQTLSLTCTVSGGSVSVTVTTVSSASTKGPSVFP", "Enzyme"],
    ["Actine", "MDMEKGKKYGGYAGRYVGGGDLAALDKYEGFFATGGQAEARKKLVDTFGDDDAQGGEL", "Mouvement"],
    ["Myosine", "SQEEIKRIIRDLTKHPNQRIYFESGVCQAGFQGSHMVEALYDVCGERGFFYTPKA", "Mouvement"],
    ["Lysozyme", "KRNGIAGWCHAARGKDIAALDLQTLKADIAAR", "Enzyme"],
    ["Récepteur Z", "HPIPIPIKQSLNSEMIKLPQNNYPFLVYYVPITQ", "Récepteur"]
]
````

**Q1.1 : Créez un dictionnaire où les clés sont les noms des protéines et les valeurs sont des dictionnaires contenant la séquence et la fonction.**

````python
# Votre code ici
````

<span style='background :red' > Correction </span>

````python
# 1. Création du dictionnaire
dict_proteines = {}
for proteine in proteines:
    dict_proteines[proteine[0]] = {"sequence": proteine[1], "fonction": proteine[2]}
print(dict_proteines)
````

**Q1.2 : À partir du dictionnaire, construisez un DataFrame Pandas où chaque ligne représente une protéine et les colonnes correspondent au nom, à la séquence et à la fonction.**

````python
# Votre code ici
````

<span style='background :red' > Note _/10 </span>

````python
import pandas as pd
df = pd.DataFrame.from_dict(dict_proteines, orient='index')
print(df)
````

**Q1.3 : Affichez le nombre de protéines pour chaque fonction biologique.**

**Q1.5 : Calculez la longueur moyenne des séquences pour chaque fonction biologique.**

````python
# Votre code ici
````

<span style='background :red' > Correction </span>

````python
moyennes_par_fonction = {}

# Parcourir le dataframe et calculer les moyennes
for index, row in df.iterrows():
    fonction = row['fonction']
    longueur = len(row['sequence'])
    if fonction in moyennes_par_fonction:
        moyennes_par_fonction[fonction].append(longueur)
    else:
        moyennes_par_fonction[fonction] = [longueur]

# Calculer les moyennes et afficher les résultats
for fonction, longueurs in moyennes_par_fonction.items():
    print(f"Fonction {fonction}: {sum(longueurs) / len(longueurs)}")
````
