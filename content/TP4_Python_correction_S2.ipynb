# Les exemples du cours

````python
# --- CLASSE MÈRE ---
class Vehicule:
    def __init__(self, marque, modele, annee, nb_roues):
        self.marque = marque
        self.modele = modele
        self.annee = annee
        self.nb_roues = nb_roues

    # La méthode commune à tous les véhicules
    def demarrer(self):
        print(f"Le véhicule {self.marque} {self.modele} s'allume en douceur.")

    def afficher_infos(self):
        print(f"Véhicule : {self.marque} {self.modele} | Année : {self.annee} | Roues : {self.nb_roues}")
````

````python
# --- CLASSES FILLES ---
class Voiture(Vehicule):
    def __init__(self, marque, modele, annee, nombre_portes=5):
        super().__init__(marque, modele, annee, nb_roues=4)
        self.nombre_portes = nombre_portes

    # Méthode spécifique aux voitures
    def ouvrir_coffre(self):
        print(f"Le coffre de la {self.modele} s'ouvre. On peut ranger les bagages !")


class Moto(Vehicule):
    def __init__(self, marque, modele, annee, type_moto="Standard"):
        super().__init__(marque, modele, annee, nb_roues=2)
        self.type_moto = type_moto

    # Surcharge : la moto démarre différemment du véhicule classique
    def demarrer(self):
        print(f"La moto {self.marque} {self.modele} démarre au quart de tour. Vroooaam !")
````

````python
# 1. La Tesla (Voiture)
tesla = Voiture("Tesla", "Model Y", 2022)
tesla.afficher_infos()
tesla.demarrer()       # Hérite du comportement de Vehicule
tesla.ouvrir_coffre()  # Utilise son comportement spécifique

print("\n" + "-" * 40 + "\n") # Séparation visuelle

# 2. La Yamaha (Moto)
yamaha = Moto("Yamaha", "MT-07", 2023)
yamaha.afficher_infos()
yamaha.demarrer()      # Utilise son propre comportement (surcharge)

# yamaha.ouvrir_coffre() --> Provoquerait une erreur car une Moto n'a pas cette méthode !
````

## 1 - Création de classe pour des séquences biologiques

**Q1.1: Créer une classe python ```Seq``` qui a pour attribut ```id```, ```sequence```. Créer ensuite l'objet ```seq1 = Seq("Myseq","ATGTCCTG")```.**

````python
# Votre code ici
````

````python
class Seq :
    
    id = ""
    sequence = ""
        
    def __init__(self, id, sequence) :
        self.id = id
        self.sequence = sequence
        
#Creation de l'objet
seq1 = Seq("Myseq","ATGTCCTG")
````

**Q1.2: Ajouter une méthode ```length()``` à Seq qui permet de donner la taille de séquence. Recréer seq1 et lancer cette méthode avec la commande ```seq1.length()```.**

````python
# Votre code ici
````

<span style='background :red' > Correction </span>

````python
class Seq :
    id = ""
    sequence = ""
    def __init__(self, id, sequence) :
        self.id = id
        self.sequence = sequence
    
    def length(self) :
        len_seq = len(self.sequence)
        return len_seq
               
# Creation de l'objet
seq1 = Seq("Myseq","ATGTCCTG")
print(seq1.length())
seq2 = Seq("Myseq2","ATGGGGGGGGGGGGGTCCTG")
print(seq2.length())
````
